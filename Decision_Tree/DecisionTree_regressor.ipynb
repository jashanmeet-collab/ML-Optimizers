```python
import numpy as np
import matplotlib.pyplot as plt
import random
from sklearn.tree import DecisionTreeRegressor
```

```python
m=50
c=200
rd_range=10000
num_points=50
def createdata(num_points):
  X=[]
  Y=[]
  #m=5
  #c=20
  #rd_range=50
  for i in range(num_points):
    X.append(i+2)
    y=m*i*i*i+m*i*i+c+20*i*random.random()*rd_range
    Y.append(y)
  return X,Y
  
```

```python
random.seed(1)
X1,Y1=createdata(num_points)
X2,Y2=createdata(num_points)
plt.scatter(X1,Y1, label = '1')
plt.scatter(X2,Y2, label = '2')
plt.legend()
plt.show()
```

```python
X_train=np.array(X1)
X_test=np.array(X2)
X_train=X_train.reshape(X_train.shape[0],1)
X_test=X_test.reshape(X_test.shape[0],1)
```

#DecisionTree Reg

```python
reg= DecisionTreeRegressor(max_depth=10,min_samples_leaf=2)
A=reg.fit(X_train,Y1)
#reg.fit(X_test,Y2)
```

```python
Y_predict_train=reg.predict(X_train)
Y_predict_test=reg.predict(X_test)
```

```python
plt.scatter(X_train,Y1,label="train")
plt.scatter(X_train,Y_predict_train,label="Trainprediction",color="brown")
plt.legend()
plt.show()
```

```python
plt.scatter(X_test,Y2,label="Test")
plt.scatter(X_test,Y_predict_test,label="Testprediction",color="red")
plt.legend()
plt.show()
```

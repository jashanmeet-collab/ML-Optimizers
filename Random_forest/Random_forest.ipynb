```python
import numpy as np
import matplotlib.pyplot as plt
import random
from sklearn.ensemble import RandomForestRegressor
```

```python
m=50
c=200
rd_range=10000
num_points=50
def createdata(num_points):
  X=[]
  Y=[]
  #m=5
  #c=20
  #rd_range=50
  for i in range(num_points):
    x_feature = i + 2*random.random()
    X.append(x_feature)
    y=m*i*i*i+m*i*i+c+20*i*random.random()*rd_range
    Y.append(y)
  return X,Y
  
```

```python
random.seed(1)
X1,Y1=createdata(num_points)
X2,Y2=createdata(num_points)
plt.scatter(X1,Y1, label = '1')
plt.scatter(X2,Y2, label = '2')
plt.legend()
plt.show()
```

```python
X_train=np.array(X1)
X_test=np.array(X2)
X_train=X_train.reshape(X_train.shape[0],1)
X_test=X_test.reshape(X_test.shape[0],1)
```

```python
reg_RF = RandomForestRegressor(random_state=2)
reg_RF.fit(X_train,Y1)
```

```python
Y_predict_train=reg_RF.predict(X_train)
Y_predict_test=reg_RF.predict(X_test)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(X_train,Y1,label="Train")
plt.scatter(X_train,Y_predict_train,label="Train Prediction",color="brown")
plt.legend(fontsize = 20)
plt.title('Random Forest', fontsize=30)
plt.xticks(fontsize= 15)
plt.yticks(fontsize= 15)

plt.show()
```

```python
plt.scatter(X_test,Y2,label="Test Ground")
plt.scatter(X_test,Y_predict_test,label="Testprediction",color="red")
plt.legend()
plt.show()
```

```python
import random 

m = 5
c = 10
noise = 400

def createdata(num_points=100):
  X = []
  Y = []
  for i in range(num_points):
    x = random.random()*100 # x is in between 0 to 100
    y = m*(x)*(x) + c + random.random()*noise
    X.append(x)
    Y.append(y)
    #print(x, y)
  return X, Y
```

```python
#random.seed(1)
X1,Y1=createdata(num_points)
X2,Y2=createdata(num_points)
plt.figure(figsize=(15, 5))

plt.plot(Y1, label = '1')
plt.plot(Y2, label = '2')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Random Forest', fontsize=30)


plt.legend()
plt.show()
```
